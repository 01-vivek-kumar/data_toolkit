```python
# Explain the following terms
#  The difference in np.array, np.asarray and np.asanyarray
#  The difference between Deep copy and shallow copy

# Answer -

# np.array - It is used to create a new NumPy array from any input like a list or tuple.
# It always makes a new copy unless we set copy=False and the input is already an array.

list1 = [2,3,4]
arr_list = np.array(list1)
print(arr_list)

# # np.asarray - Converts the input to a NumPy array, but it avoids making a new copy if the input is already a NumPy array.
# arr_list2 = np.asarray(arr_list)

# print(arr_list2 is arr_list) - True

# #np.asanyarray - Same as np.asarray, but it allows returning subclasses like np.matrix or masked arrays if the input is already one.
# arr3 = np.asanyarray(arr_list)
# print(arr3 is arr_list)


# Shallow Copy:

# Creates a new object but doesn't copy the actual elements. Instead, it references the original data.
# Changes in the copied object can affect the original object if it’s mutable

a = arr_list
a[0] = 100
print(a)

#Deep copy - Creates a new object and recursively copies all the elements, making it completely independent of the original object.
# Changes in the copied object don’t affect the original.
arr_shallow = arr_list.copy()
arr_shallow[1] = 100
print(arr_shallow)
```

```python
import bokeh.io
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
x = np.linspace(1,5, 10)  # 100 points from 0 to 4*pi
y = np.sin(x)  # Sine function values for each x point
```

```python

p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
p.grid.grid_line_dash = [6,4 ]
```

```python
# Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
# the title as 'Interactive Pie Chart'.
import plotly.graph_objects as go
import numpy as np

# Generate random data for the pie chart
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)

# Create the pie chart
fig = go.Figure(data=[go.Pie(labels=categories, values=values, textinfo='label+percent', pull=[0.1, 0, 0, 0, 0])])

# Add the title
fig.update_layout(title='Interactive Pie Chart')

# Show the pie chart
fig.show()
```
